In this notebook descriptive statistics are calculated for both the outcome variables (eighteen drug types) and the independent variables (twelve sociodemographic/personality variables). Tests of normality are also performed for more information, in case regression methods are used.

```python
#import libraries
import pandas as pd
from scipy import stats
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
#read in drug consumption file
df = pd.read_csv("../drug_consumption_cap_20230505.csv")
df
```

```python
#check columns
df.columns
```

## Descriptives for Independent Variables

```python
#create descriptives
input_descriptives = df.iloc[:,:13].describe().transpose()
input_descriptives
```

```python
#look at columns
df.iloc[:,:13].columns
```

```python
#test shapiro statistic for normality
stats.shapiro(df.iloc[:,0])
```

```python
#create final table of descriptives for the independent variables 
normal_test=[]

#create loop
for i in df.iloc[:,:13].columns:
    #retieve shapiro statistic
    sw_stat = stats.shapiro(df[i])[0]
    #retrieve shapiro p-value
    sw_p_val = stats.shapiro(df[i])[1]
    #retrieve D'Agostino statistic 
    DA_stat = stats.normaltest(df[i])[0]
    #retrieve D'Agostino statistic p-value
    DA_p_val = stats.normaltest(df[i])[1]
    #encode with thresholds for normality
    result_sw = ["not normal" if sw_p_val < 0.0500 else "normal"][0]
    result_da = ["not normal" if DA_p_val < 0.0500 else "normal"][0]
    #create dictionary for row
    row = {"shapiro_stat":sw_stat, "shapiro_pval":sw_p_val, "SW_result":result_sw, "dago_stat":DA_stat, "dago_pval":DA_p_val, "DA_result":result_da}
    #append to list
    normal_test.append(row)
    #print(row)
#create final dataframe
norm_df = pd.DataFrame(normal_test, index = df.iloc[:,:13].columns)    
#display df
norm_df
```

```python
final_input_descrip = input_descriptives.merge(norm_df, left_index=True, right_index=True)
```

```python
final_input_descrip
```

```python
#use loop to plot histograms and qq plots for all independent variables
for i in df.iloc[:,:13].columns:
    #print name of variable
    print(("-")*40+i+("-")*40)
    
    #set seaborn theme
    sns.set_theme()
    
    #plot histogram with kde
    sns.displot(data= df[i],kde=True,height=5,aspect=1.5)
    
    #set figure properties
    fig, ax = plt.subplots(figsize=(12,8))
    
    #QQ plot
    stats.probplot(df[i], dist="norm", plot=plt)
    plt.show()
```

## Outcome Variable Descriptives

```python
#retreive only the outcome druge/alcohol variables
describe_outcomes = df.iloc[:,13:].describe().transpose()
describe_outcomes
```

```python
#Get counts for levels in each drug variable
df.iloc[:,13:].apply(pd.Series.value_counts)
```

```python
#tranpose table
cate_value_counts = df.iloc[:,13:].apply(pd.Series.value_counts).transpose()
cate_value_counts
```

```python
#create percentage distribution for consumption levels for each drug
t = pd.concat([df[column].value_counts(normalize=True) for column in df.iloc[:,13:]], axis = 1).transpose()
```

```python
t["totals"]= t.sum(axis=1)
```

```python
#select only the final classes and total needed
pct_descript = t[["CL0","CL1","CL2","CL3","CL4","CL5","CL6",'totals']]
pct_descript
```

```python
#final export to write to an Excel Notebook
with pd.ExcelWriter("descriptives.xlsx") as writer:
    final_input_descrip.to_excel(writer,sheet_name='input_var_summary')
    describe_outcomes.to_excel(writer,sheet_name='ouput_var_summary')
    cate_value_counts.to_excel(writer,sheet_name='distribution_output')
    pct_descript.to_excel(writer,sheet_name='pct_distributions')
```
